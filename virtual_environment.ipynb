```python
import matplotlib.pyplot as plt
import numpy as np

# --- Constants and Configuration ---
num_computers = 30
power_per_computer = 150  # in watts
lighting_power_density = 10  # W/m²
area = 100  # Room area in m²
indoor_temp = 24  # Constant indoor temperature in °C
cooling_factor = 5  # W/m²/°C
hours = np.arange(8, 19)  # 8 AM to 6 PM

# --- Power Calculations ---
# Computers
power_computers_kw = (num_computers * power_per_computer) / 1000  # in kW

# Lights
power_lights_kw = (lighting_power_density * area) / 1000  # in kW

# Simulated outside temperature from 25°C to 35°C during the day
outside_temp = np.linspace(25, 35, len(hours))
delta_temp = outside_temp - indoor_temp

# Cooling power (in kW) per hour
cooling_power_kw = (delta_temp * cooling_factor * area) / 1000

# Total hourly power usage (in kW)
total_power_kw = power_computers_kw + power_lights_kw + cooling_power_kw

# Since each time step is 1 hour, energy per hour = power
energy_per_hour_kwh = total_power_kw
total_energy_kwh = np.sum(energy_per_hour_kwh)

# --- Plotting ---
plt.figure(figsize=(12, 6))
plt.plot(hours, energy_per_hour_kwh, marker='o', label='Total Power Usage (kWh)')
plt.plot(hours, cooling_power_kw, linestyle='--', label='Cooling Power Only (kWh)', color='skyblue')
plt.hlines(power_computers_kw + power_lights_kw, 8, 18, colors='gray', linestyles='dotted', label='Computers + Lights')
plt.title(f'Energy Consumption Simulation (Total: {total_energy_kwh:.2f} kWh)')
plt.xlabel('Hour of Day')
plt.ylabel('Energy Usage (kWh)')
plt.xticks(hours)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# --- Optional: Print summary ---
print(f"Total Energy Consumption from 8 AM to 6 PM: {total_energy_kwh:.2f} kWh")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# --- Constants and Configuration ---
num_computers = 30
power_per_computer = 150  # in watts
lighting_power_density = 10  # W/m²
area = 100  # Room area in m²
indoor_temp = 24  # Constant indoor temperature in °C
cooling_factor = 5  # W/m²/°C
hours = np.arange(8, 19)  # 8 AM to 6 PM

# --- Power Calculations ---
# Computers
power_computers_kw = (num_computers * power_per_computer) / 1000  # in kW

# Lights
power_lights_kw = (lighting_power_density * area) / 1000  # in kW

# Simulated outside temperature from 25°C to 35°C
outside_temp = np.linspace(25, 35, len(hours))
delta_temp = outside_temp - indoor_temp

# Cooling power (in kW)
cooling_power_kw = (delta_temp * cooling_factor * area) / 1000

# Total hourly power usage (in kW)
total_power_kw = power_computers_kw + power_lights_kw + cooling_power_kw

# Energy per hour = power for each hour
energy_per_hour_kwh = total_power_kw
total_energy_kwh = np.sum(energy_per_hour_kwh)

# --- Data Table ---
df = pd.DataFrame({
    'Hour': hours,
    'Outside Temp (°C)': outside_temp.round(2),
    'Cooling Power (kWh)': cooling_power_kw.round(2),
    'Total Energy (kWh)': energy_per_hour_kwh.round(2)
})

print("Hourly Energy Consumption:\n")
print(df)

# --- Pie Chart of Total Energy Usage ---
total_computers_energy = power_computers_kw * len(hours)
total_lights_energy = power_lights_kw * len(hours)
total_cooling_energy = np.sum(cooling_power_kw)

labels = ['Computers', 'Lights', 'Cooling']
sizes = [total_computers_energy, total_lights_energy, total_cooling_energy]
colors = ['orange', 'yellow', 'skyblue']

plt.figure(figsize=(6, 6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Energy Distribution (8 AM to 6 PM)')
plt.axis('equal')
plt.tight_layout()
plt.show()

# --- Line Plot ---
plt.figure(figsize=(12, 6))
plt.plot(hours, energy_per_hour_kwh, marker='o', label='Total Power Usage (kWh)')
plt.plot(hours, cooling_power_kw, linestyle='--', label='Cooling Power Only (kWh)', color='skyblue')
plt.hlines(power_computers_kw + power_lights_kw, 8, 18, colors='gray', linestyles='dotted', label='Computers + Lights')
plt.title(f'Energy Consumption Simulation (Total: {total_energy_kwh:.2f} kWh)')
plt.xlabel('Hour of Day')
plt.ylabel('Energy Usage (kWh)')
plt.xticks(hours)
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()

# --- Summary ---
print(f"\nTotal Energy Consumption from 8 AM to 6 PM: {total_energy_kwh:.2f} kWh")
```
